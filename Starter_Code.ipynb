## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE

updated_app_df = application_df[["APPLICATION_TYPE",	"AFFILIATION",	"CLASSIFICATION",	"USE_CASE",	"ORGANIZATION"	,"STATUS"	,"INCOME_AMT"	,"SPECIAL_CONSIDERATIONS"	,"ASK_AMT",	"IS_SUCCESSFUL"]]
updated_app_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
application_types_to_replace = ['T17','T15','T29','T14','T25','T2','T12','T13','T9']
# Replace in dataframe
for app in application_types_to_replace:
    updated_app_df['APPLICATION_TYPE'] = updated_app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
updated_app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

classifications_to_replace = ['C7000','C1700',
 'C4000',
 'C5000',
 'C1270',
 'C2700',
 'C2800',
 'C7100',
 'C1300',
 'C1280',
 'C1230',
 'C1400',
 'C7200',
 'C2300',
 'C1240',
 'C8000',
 'C7120',
 'C1500',
 'C1800',
 'C6000',
 'C1250',
 'C8200',
 'C1238',
 'C1278',
 'C1235',
 'C1237',
 'C7210',
 'C2400',
 'C1720',
 'C4100',
 'C1257',
 'C1600',
 'C1260',
 'C2710',
 'C0',
 'C3200',
 'C1234',
 'C1246',
 'C1267',
 'C1256',
 'C1370',
 'C1248',
 'C6100',
 'C1820',
 'C1900',
 'C1236',
 'C3700',
 'C2570',
 'C1580',
 'C1245',
 'C2500',
 'C1570',
 'C1283',
 'C2380',
 'C1732',
 'C1728',
 'C2170',
 'C4120',
 'C8210',
 'C2561',
 'C4500',
 'C2150',
 'C2190',
 'C4200',
 'C5200',
 'C2600']

# Replace in dataframe
for cls in classifications_to_replace:
    updated_app_df['CLASSIFICATION'] = updated_app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
updated_app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE

ap_dummies = pd.get_dummies(updated_app_df)
ap_dummies
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
X = ap_dummies.drop(["IS_SUCCESSFUL"],axis=1)
y = ap_dummies["IS_SUCCESSFUL"]


# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
feature = X_train_scaled.shape[1]
hide_node1=15
hide_node2=14
hide_node3=1

nn = tf.keras.models.Sequential()










# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hide_node1, activation="relu", input_dim=feature))
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hide_node2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=hide_node3, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=200)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
